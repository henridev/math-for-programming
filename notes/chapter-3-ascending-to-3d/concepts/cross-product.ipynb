- similar to dot product in that the result is determined by magnitude and direction of inputs
- different in that output also has a direction
- cross product shoudl tell us something about orientation
- cross product is perpendicular to the two input vectors


- dot product tells us how perpendicular two vectors are
- cross product length tels us what the area between to vectors is

- cross product length will be greatest if vectors are perpendicular

```python
import sys
sys.path.insert(0,'..')
import numpy as np
from vectors import *;
from draw3d import *;
from draw2d import *;
```

```python
def cross(u, v):
    ux,uy,uz = u
    vx,vy,vz = v
    return (uy*vz - uz*vy, uz*vx - ux*vz, ux*vy - uy*vx);
```

```python
u = (1,4,0)
v = (4,2,0)

# index to u middle to v
cross_prod_1 = cross(u,v)
# index to v middle to u
cross_prod_2 = cross(v,u)


draw3d(
    Arrow3D(u, color=green),
    Arrow3D(v, color=blue),
    Arrow3D(cross_prod_1, color=purple),
    Arrow3D(cross_prod_2),
);
```

```python


# with two vectors we can cover an entire plain in 3d space

u = (3,4,2)
v = (4,-2,3)

u_options = [(u[0] * i, u[1] * i, u[2] * i) for i in np.arange(-10,10, 0.1)]
v_options = [(v[0] * i, v[1] * i, v[2] * i) for i in np.arange(-10,10, 0.1)]

result = [add(*[u,v]) for (u, v) in zip(u_options, v_options)]
result_2 = [add(*[u,v]) for (u, v) in zip(u_options, reversed(v_options))] 

points_u = [Points3D(pnt, color=green) for pnt in u_options]
points_v = [Points3D(pnt, color=blue) for pnt in v_options]
points_result = [Points3D(pnt, color=red) for pnt in result]
points_result_2 = [Points3D(pnt, color=red) for pnt in result_2]

surface = Polygon3D(*[
    add(*[u_options[0],v_options[-1]]),
    add(*[u_options[0],v_options[0]]),
    add(*[u_options[-1],v_options[0]]),
    add(*[u_options[-1],v_options[-1]]),
  ],
  color=purple
)



draw3d(
    *points_u,
    *points_v,
    surface
);
```

```python
from math import *;
from plot3dcube import *;

# will be equal to length of cross product
def area_spanned_by_vectors(v1, v2, angle, inRadians = True):
    angle = angle if inRadians else radians(angle)
    # we can see the sin of an angle as a sort of ratio getting smaller
    # together with the angle
    return length(v1) * length(v2) * sin(angle)


u = (2, 0, 0)
v = (0, 2, 0)

cross_product = cross(u ,v)
plot3dcube(u, v, (0,0,0))

draw3d(
    Arrow3D(u),
    Arrow3D(v),
    Arrow3D(cross_product),
);

print(
    area_spanned_by_vectors(u, v, 90, inRadians=False),
    length(cross_product)
)


u = (0, 2, 0)
v = (0, 0, -2)

cross_product = cross(u ,v)
plot3dcube(u, v, (0,0,0))

draw3d(
    Arrow3D(u),
    Arrow3D(v),
    Arrow3D(cross_product),
);

print(
    area_spanned_by_vectors(u, v, 90, inRadians=False),
    length(cross_product)
)
```
